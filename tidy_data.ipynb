# Attendance Data

# Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
# read the attendance table
url = env.get_url(env.user, env.host, env.password, 'tidy_data')

df = pd.read_sql('SELECT * FROM attendance', url)
df.head()
```

```python
#rename the first column
df.rename(columns = {'Unnamed: 0' : 'name'}, inplace=True)
df.head()
```

```python
# melt dataframe to obtain one column of values
df = df.melt(id_vars=['name'], var_name='date', value_name='attendance')
df
```

```python
# change attendance values to numeric
df.attendance = df.attendance.map({"P" : 1, "H" : .5, "T" : .9, "A" : 0})
df
```

```python
# groupby name and average the attendance values
df.groupby(['name']).mean()
        
```
